# Density Matrix from Noise Circuit in Qiskit

```python
from qiskit import QuantumCircuit
import qiskit.quantum_info as qi
from qiskit_aer import AerSimulator
from qiskit_aer.noise import NoiseModel, mixed_unitary_error
```

```python
l = qi.DensityMatrix.from_label('l')
r = qi.DensityMatrix.from_label('r')
p = qi.DensityMatrix.from_label('+')

rho = 8/10*r + 1/10*l + 1/10*p
rho.draw("latex")
```

```python
S = qi.Operator.from_label('S')
I = qi.Operator.from_label('I')
Sdg = S.adjoint()

prob_S = 0.8
prob_Sdg = 0.1
prob_I = 0.1

S_noisy = mixed_unitary_error([(S, prob_S), (Sdg, prob_Sdg), (I, prob_I)])
print(S_noisy)
```

```python
noise_m = NoiseModel()
noise_m.add_quantum_error(S_noisy,['hello'],[0])
print(noise_m)
```

```python
qc = QuantumCircuit(1)
qc.h(0)
qc.id(0)
qc.save_density_matrix()
qc.draw()
```
